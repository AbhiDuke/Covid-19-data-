```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import math
```

```
#dataset of Cases
df = pd.read_csv("/dbfs/FileStore/tables/time_series_covid19_confirmed_US.csv", header='infer')
df.head()
```

```
def clean_confirmed():
  df_1 = df.drop(labels=['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Province_State','Country_Region', 'Lat', 'Long_','Combined_Key'],axis = 1)
  no_of_cases={}
  for i in df_1:
    no_of_cases[i]=sum(df_1[i]) 
  new_data = pd.DataFrame([no_of_cases.keys(),no_of_cases.values()])
  df_2 = new_data.T
  df_2.columns = ['Date','Cases']
  df_2['Date']=pd.to_datetime(df_2['Date'])
  return df_2
df_2 = clean_confirmed()
print(df_2.head())
```

```
#data set #deaths JHU
f = "/dbfs/FileStore/tables/time_series_covid19_deaths_US_1_-f2823.csv"
def deaths_fun(f):
  df_deaths = pd.read_csv(f)
  df_deaths_1=df_deaths.drop(labels=['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Province_State','Country_Region', 'Lat', 'Long_','Combined_Key','Population'], axis = 1)
  no_of_deaths={}
  for i in df_deaths_1:
    no_of_deaths[i]=sum(df_deaths_1[i]) 
  new_data_1 = pd.DataFrame([no_of_deaths.keys(),no_of_deaths.values()])
  df_deaths_2 = new_data_1.T
  df_deaths_2.columns = ['Date','Deaths']
  df_deaths_2['Date']=pd.to_datetime(df_deaths_2['Date'])
  return df_deaths_2
df_dead= deaths_fun(f=f)
print(df_dead.head())
```

```
states_l = ["Alaska", "Alabama", "Arkansas", "American Samoa", "Arizona", "California", "Colorado", "Connecticut", "District ", "of Columbia", "Delaware", "Florida", "Georgia", "Guam", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Puerto Rico", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Virgin Islands", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
 
```

```
# Cleaning mobility data to data only valid to USA.
mobi_list = []
for i in range(0,len(mob_df)):
  if mob_df['region'][i] == 'United States' or mob_df['alternative_name'][i] == 'USA':
    mobi_list.append(mob_df.iloc[i])
  elif mob_df['sub-region'][i] in states_l:
    mobi_list.append(mob_df.iloc[i])  
  elif mob_df['country'][i] == 'United States':
    mobi_list.append(mob_df.iloc[i])
  else:
    continue
mobi_data = pd.DataFrame(mobi_list)
nn = mobi_data.reset_index()
mobi_data.head()
```

```
#Filtering data with respect to transportation mode.

driving = []
walking = []
transit = [] 
for i in range(0,len(nn)):
  if nn['transportation_type'][i]=='driving':
    driving.append(nn.iloc[i])
  elif nn['transportation_type'][i]=='walking':
    walking.append(nn.iloc[i])
  else:
    transit.append(nn.iloc[i])
drive_df = pd.DataFrame(driving)
walk_df = pd.DataFrame(walking)
transit_df = pd.DataFrame(transit)    
print(drive_df.head())
```

```
# Getting avg transportation score for each day in USA (Country wide analysis)
def d_mobi(data):
  mobility_percent = {}
  df = data.drop(labels=[ 'index','geo_type', 'region', 'transportation_type',
       'alternative_name', 'sub-region', 'country'],axis = 1)
 # print(df)
  for i in df:
    mobility_percent[i] = df[i].mean()
  form = pd.DataFrame([mobility_percent.keys(),mobility_percent.values()])  
  form = form.T
  return form
driving_data = d_mobi(data=drive_df)
walking_data = d_mobi(data=walk_df)
tran_data = d_mobi(data=transit_df)
driving_data.columns = ['Date', 'Drive_rate']
driving_data['Date'] = pd.to_datetime(driving_data['Date'])
walking_data.columns = ['Date', 'Walking_rate']
walking_data['Date'] = pd.to_datetime(walking_data['Date'])
tran_data.columns = ['Date', 'transit_rate']
tran_data['Date'] = pd.to_datetime(tran_data['Date'])
print(driving_data.head())
```

```
#final data together combining mobility with cases and deaths.
final_df = pd.merge(df_2,df_dead)
final_df = pd.merge(final_df,driving_data)
final_df=pd.merge(final_df,walking_data)
final_df=pd.merge(final_df,tran_data)
final_df.head()
```

```
#Extracting Cases per day and deaths per day from cummilative cases and deaths
Cases_perday=[]
Deaths_day = []
for i in range(0,len(final_df['Cases'])):
  if i == 0:
    Cases_perday.append(final_df['Cases'][i])
    Deaths_day.append(final_df['Deaths'][i])
  else:
    Cases_perday.append(final_df['Cases'][i] - final_df['Cases'][i-1])
    Deaths_day.append(final_df['Deaths'][i]-final_df['Deaths'][i-1])
final_df['Deaths_day'] = Deaths_day
final_df['Cases_perday'] = Cases_perday
final_df.head()
```

```
Y = data_prep['Cases_perday']
X = data_prep.drop(labels=['Cases_perday'],axis = 1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,train_size = 0.25, random_state = 1)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train,y_train)
predictions = model1.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print('MODEL Linear Regressioon')
print('MAE : ',mean_absolute_error(y_test,predictions))
print('MSE : ',mean_squared_error(y_test,predictions)) 
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions)))
print('R^2 : ',explained_variance_score(y_test,predictions))
```

```
from sklearn.linear_model import ElasticNet,Lasso, Ridge

model2 = ElasticNet()


model2.fit(X_train,y_train)
predictions_2 = model2.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print("MODEL ElasticNet")
print('MAE : ',mean_absolute_error(y_test,predictions_2))
print('MSE : ',mean_squared_error(y_test,predictions_2)) 
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions_2)))
print('R^2 : ',explained_variance_score(y_test,predictions_2))
```

```
model3 = Lasso(max_iter=100000)
model3.fit(X_train,y_train)
predictions_3 = model3.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print("MODEL Lasso Regression")
print('MAE : ',mean_absolute_error(y_test,predictions_3))
print('MSE : ',mean_squared_error(y_test,predictions_3)) 
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions_3)))
print('R^2 : ',explained_variance_score(y_test,predictions_3))
```

```
model4 = Ridge()
model4.fit(X_train,y_train)
predictions_4 = model4.predict(X_test)
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print("MODEL Ridge Regression")
print('MAE : ',mean_absolute_error(y_test,predictions_4))
print('MSE : ',mean_squared_error(y_test,predictions_4)) 
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions_4)))
print('R^2 : ',explained_variance_score(y_test,predictions_4))
```

```
params = {'n_estimators': 500,
          'max_depth': 4,
          'min_samples_split': 5,
          'learning_rate': 0.01,
          'loss': 'ls'}
from sklearn.ensemble import GradientBoostingRegressor
reg = GradientBoostingRegressor(**params)
reg.fit(X_train, y_train)
predictions_5 = reg.predict(X_test)

mse = mean_squared_error(y_test, reg.predict(X_test))

from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
print("Gradient Bossting Reg")
print('MAE : ',mean_absolute_error(y_test,predictions_5))
print('MSE : ',mean_squared_error(y_test,predictions_5)) 
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions_5)))
print('R^2 : ',explained_variance_score(y_test,predictions_5))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```
