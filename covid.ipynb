```python
import json
import pandas as pd
```

```python
file = open('covid19.json','r')
```

```python
data = json.load(file)
```

```python
new_dic1={}
for country in data:
    if country not in new_dic1:
        new_dic1[country]={}
    for list_cases in data[country]:
        if list_cases['deaths']!= 0 : 
            new_dic1[country]['deaths']=list_cases['date']
            break   
    for list_cases in data[country]:
        if list_cases['confirmed']!= 0: 
            new_dic1[country]['confirmed']=list_cases['date']
            break
    for list_cases in data[country]:
        if list_cases['recovered']!= 0: 
            new_dic1[country]['recovered']=list_cases['date']        
            break
```

```python
new_dic1
```

```python
df = pd.DataFrame.from_dict(new_dic1,orient='index')
```

```python
df['country']=df.index
df.reset_index(inplace=True)
```
